### 제너레이터

- 루프의 반복 동작을 제어할 수 있는 루틴 형태
- yield 키워드 사용
    : return과 달리 함수가 종료되지 않음


```python
def get_natural_number():
    n = 0
    while True:
        n+=1
        yield n

g = get_natural_number()

for _ in range(0,100):
    print(next(g), end=' ', flush = True)

print()
print(next(g))  # 기존 값이 저장되어 있는 상태임. n=0으로 초기화되지 않음. g는 글로벌하게 선언되어 있기 때문
```

### enumerate()

- 여러가지 enumerable한 자료형 (list, set, tuple 등)을, 인덱스를 포함한 enumerate 객체로 리턴한다.



```python
a = list(range(5))

print(a)

b = enumerate(a)

print(type(b))
print(b)

print(list(b))
```

```python
a = ['a1', 'b2', 'c3']

for (i,v) in enumerate(a):
    print(i,v)
```

```python
idx = 1
fruit = "Apple"

# How to print " 2 : Apple"

## version 1 : curly brace with index
print ('{0} : {1}'.format(idx +1, fruit))

## version 2 : curly brace without index
print ('{} : {}'.format(idx +1, fruit))

##version 3 : f-string (formatted string literal, 파이썬 3.6 이상에서만 지원)
print(f'{idx+1} : {fruit}')

```

### pass
- 코드 전체의 골격 잡아놓고 (목업) 내부적으로 차근차근 접근할 때, 인덴트 오류 방지를 위함

```python
class MyClass(object):
    def method_a(self):
        pass
    def method_b(self):
        print("Method B")

c= MyClass()

print(c.method_b())
```
